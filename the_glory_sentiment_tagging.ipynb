```python
for i in range(10, 11):
    # if i == 10:  # 10화가 아직 없음 10화 생기면 없애도 됨
    #     continue
    df = data_load(i)
    print(df.shape)
    target_df = df.loc[ ~df['말하는 사람'].isnull(), :]
    print(target_df.shape)
    # idx_list = df.index
    # print(idx_list)

    sentiment_list = []
    for subtitle in target_df['Subtitle'].apply(lambda x: x.split(']')[1:]):
        time.sleep(3)
        sentiment_clf = openai.ChatCompletion.create(
            model="gpt-3.5-turbo",
            messages=[{"role": "user",
                       # "content": "다음 문장에 대해 긍정, 중립, 부정 셋 중 하나로 감정분류 해줘. '신이 보기에 건방지다나? 그래서 그런지 그 꽃은 밤에만 향기가 나요 -'",
                       "content": "Please say positive, neutral or negative only: \"{}\"".format(subtitle)
                       }]
        )
        print(sentiment_clf)
        sentiment_list.append(sentiment_clf.choices[-1].message.content)
    target_df['sentiment'] = sentiment_list
    target_series = target_df['sentiment'].str.lower().str.split('.').str[0]
    df = df.merge(target_series, how='outer', left_index=True, right_index=True)
    data_save(df, i)
```
