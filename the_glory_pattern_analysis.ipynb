```python
df_ss_ep.plot(figsize=(20, 12))
plt.axhline(y=0, color='black')
plt.title('Subgraph Significance by Episode', y=1.05, fontsize=20)
plt.xticks(np.arange(16), df_ss_ep.index)
plt.tight_layout()
# plt.savefig('./output_dir/subgraph_significance_by_episode.png')
plt.show()
```

```python
import seaborn as sns
```

```python
# 각 episode에 대한 색상값 설정
# colors = ['#EF476F', '#FFD166', '#06D6A0', '#118AB2', '#073B4C', '#FF7F50', '#6495ED', '#FF1493',
#           '#B22222', '#9ACD32', '#FF00FF', '#7B68EE', '#2F4F4F', '#1E90FF', '#00FFFF', '#008000']
colors = sns.color_palette('bright', n_colors=16)
# # 각 episode에 대한 선 스타일 설정
# linestyles = ['-', '--', '-.', ':', '-', '--', '-.', ':', '-', '--', '-.', ':', '-', '--', '-.', ':']

# 각 episode에 대한 선 스타일 설정
linestyles = ['-', '--', '-.', ':', (0, (1, 1)), (0, (5, 1)), (0, (3, 1, 1, 1)), (0, (3, 5, 1, 5)), '-', '--', '-.', ':', (0, (1, 5)), (0, (3, 1, 1, 1, 1, 1)), (0, (3, 5, 1, 5, 1, 5)), (0, (1, 1, 1, 1, 1, 5))]


plt.figure(figsize=(20, 12))
# 색상값과 선 스타일을 함께 적용하여 그래프 그리기
for i, (col, ls) in enumerate(zip(colors, linestyles)):
    plt.plot(df_ss_ep.iloc[:, i], color=col, linestyle=ls, label=df_ss_ep.columns[i])

plt.axhline(y=0, color='black')
plt.title('Subgraph Significance by Episode', y=1.05, fontsize=20)
plt.xticks(np.arange(16), df_ss_ep.index)
plt.legend()
plt.tight_layout()
# plt.savefig('./output_dir/subgraph_significance_by_episode_v2.png')
plt.show()
```
